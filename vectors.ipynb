adding and subtracting vectors is an element-wise operation <br>
both the vectors should have the same dimensions <br>

```python
v = np.array([1, 5])
w = np.array([-4, 2])
u = v + w

plt.plot([0, v[0]], [0,v[1]], label='v')
plt.plot([0, w[0]], [0, w[1]], label='w')
plt.plot([0, u[0]], [0, u[1]], label='u = v + w')
plt.plot([v[0], v[0] + w[0]], [v[1], v[1] + w[1]], '--', label='v + w')
plt.plot([w[0], w[0] + v[0]], [w[1], w[1] + v[1]], '--', label='w + v')

plt.plot([0, 0], [-1, 8], 'k--')
plt.plot([-5, 2], [0, 0], 'k--')
plt.grid()
plt.legend()
```

scalar vector multiplication: a single nummeric value that scales the vector without changing the direction <br>

```python
v = np.array([1, 2])
s = 2
l = -1.2
o = 0

# scalar s * vector v; if scalar > 1 : maintains the same direction; scalar < 1: reverses the direction
w = s * v
x = l * v
z = o * v

plt.plot([-2, 2], [0, 0], 'k--')
plt.plot([0, 0], [-3, 5], 'k--')
plt.plot([0, v[0]], [0, v[1]], linewidth='5', label='v')
plt.plot([0, w[0]], [0, w[1]], label='v * 2')
plt.plot([0, x[0]], [0, x[1]], label='v * -1.2')
plt.plot([0, z[0]], [0, z[1]], label='v * 0')
plt.grid()
plt.legend()
plt.show()
```

outerproduct: multiplies the outer elements of the vectors <br>
v[mx1] w[1xn] = x[mxn] (gives a vector)
vector in v is scaled by each value in w and generates one column for every scaled vector <br>
vector w is repeated in rows with number of elements in v, and each row is scaled by the equivalent index of the element in v

```python
v = np.random.randn(5, 1)
w = np.random.randn(1, 3)

X = np.outer(v, w)
Y = np.zeros((5, 3))

# each column in Y is scaled version of vector v with the respectived indexed element in w
for i in range(3):
    Y[:, i] = v[:, 0] * w[0, i]


print(Y)
print(X)
```

dot product or scalar product: <br>
    - gives out a single number or a scalar value <br>
    - multiply each element of the first vector transposed with the respective element of the second vector and take the sum of all those resulting products <br>
    - two vectors in the dot product must be of the same size <br>
    - distribute and commutative, but not associtive <br>

dot(v, w) = s <br>
dot(v, dot(w, u)) is not valid as dot(w, u) gives a scalar and there is no dot produt between a scalar and a vector
dot(v, w) == dot(w, v)

```python
m = 4
v = np.random.randn(m, 1)
w = np.random.randn(m, 1)
x = np.random.randn(m, 1)
dot1a = np.dot(v.T, w)
dot1b = np.dot(w.T, v)
dot2 = 0
for i in range(m):
    dot2 += v[i] * w[i]

dot3 = np.sum(v * w)

# matrix multiplication of a (1, 4) and (4, 1) vector gives a single value, which for vectors can also be represented as a dot product
dot4 = np.matmul(v.T, w)

print(dot1a)
print(dot1b)
print(dot2)
print(dot3)
```

matrix multiplication can be inferred as series of dot products on respective rows and columns of two matrices: <br>
for matrix A x B = C: C[i,j] = dot(A[i, :], B[:, i])

```python
m = 3
n = 4

A = np.random.randn(m, n)
B = np.random.randn(n, m)
C = np.matmul(A, B)
D = np.zeros((m, m))

# multiplying each row of A with respective column of B to create a matrix product
for i in range(m):
    for j in range(m):
        D[i, j] = np.dot(A[i, :], B[:, j])


print(C)
print(D)
```

```python
m = 4
v = np.random.randn(m, 1)
w = np.random.randn(m, 1)
x = np.random.randn(m, 1)

# commutative
dotvw = np.dot(v.T, w)
dotwv = np.dot(w.T, v)

print(dotvw == dotwv)

# distribute v(w+x) = vw + vx
dotd = np.dot(v.T, (w + x))
dotd2 = np.dot(v.T, x) + np.dot(v.T, w)

print(dotd)
print(dotd2)
```

```python
# scaled multiplication of dot product and multiplying it with a scalar value doesn't change the sign of the dot product of the vector

v = np.random.randn(1, 3)
w = np.random.randn(3, 1)
x0 = np.cos(0)
x90 = np.cos(np.pi / 2)
x180 = np.cos(np.pi) 
x270 = np.cos(np.pi * 3/2)

dot = np.dot(v, w)
dot0 = np.dot(v*x0, w*x0)
dot90 = np.dot(v*x90, w*x90)
dot180 = np.dot(v*x180, w*x180)
dot270 = np.dot(v*x270, w*x270)

print(f"dot(v,w): {dot}")
print(f"dot(v,w) scaled by cos(0): {dot0}")
print(f"dot(v,w) scaled by cos(90): {dot90}")
print(f"dot(v,w) scaled by cos(180): {dot180}")
print(f"dot(v,w) scaled by cos(270): {dot270}")
```

norm or magnitude or length of a vector (v): ||v|| = sqrt(x^2 + y^2), where x and y are corresponding coordinates <br>
    - the norm can be derived from pythogorean theorm of a right triangle <br>
    - iimagine the vector forming a right angle triangle from the head of the vector to one of the axis <br>
    - thus taking the vector itself as a hypotenuse and that squared is equal to sum of the squares of the other two sides <br>
    - it can be further extended into higher dimensions, first deriving the norm of the two dimension vector, taking it as plane onto which the higehr dimensional vectors are projected <br>
    - the dot product between two vectors also gives the same values, upon which taking the sqrt(dot) also gives the same result <br>

```python
v = np.random.randn(1, 2)
normv = np.linalg.norm(v)
normv2 = np.sqrt(v[0][0] ** 2 + v[0][1] ** 2)
normv3 = np.sqrt(np.dot(v, v.T))


print(normv)
print(normv2)
print(normv3)
```

dot product: magitude of two vectors * cos(x); x: angle between the two vectors
if the vectors lie on the same line: cos(0) = 1; dot = ||v|| ||w||
orthogonal (90 degree or pi/2): cos(90) = 0; dot = 0
respectively, the dot product shifts in signs as it moves between 0 and 2pi

```python
# dot(a, b) = cos(ab) ||v|| ||w||
v = np.array([1, 2])
w = np.array([3, 5])

d = np.dot(v, w)
vnorm = np.linalg.norm(v)
wnorm = np.linalg.norm(w)

cosvw = np.arccos(np.dot(v, w) / (vnorm * wnorm))
print(d)
print(np.cos(cosvw) * vnorm * wnorm) 
```

Cauchy-Schwarz inequality: abs(v.T w) <= ||v|| ||w|| <br>
dot(v, w) = ||v|| ||w|| cos(vw); since cos(x) is always between -1 and 1, the dot product will be ||v|| ||w|| if angle is 0 and if it not, it will be multiplied by a value which is less than 1 and thus scaling down the value of ||v|| and ||w|| <br>

||v|| ||w|| cos(vw) <= ||v|| ||w||

```python
v = np.random.randn(1, 4)
w = np.random.randn(4, 1)

dot = np.abs(np.dot(v, w))
vnorm = np.linalg.norm(v)
wnorm = np.linalg.norm(w)

wvnorm = vnorm * wnorm
print(dot)
print(wvnorm)
```
